```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# MiniGrid 6x6

```python
df = pd.read_csv("q_learning_MiniGrid-Empty-6x6-v0_10000.csv")
rew_dcrm = df.avg_reward.values
std_dcrm = np.array([0])

b = 100
for i in range(1,len(rew_dcrm)):
  if i-b<0:
    std_dcrm = np.append(std_dcrm, np.std(rew_dcrm[0:i])) 
  else:
    std_dcrm = np.append(std_dcrm, np.std(rew_dcrm[i-b:i])) 


df = pd.read_csv("qrm_learning_MiniGrid-Empty-6x6-v0_10000.csv")
rew_ddcrm = df.avg_reward.values
std_ddcrm = np.array([0])
for i in range(1,len(rew_ddcrm)):
  if i-b<0:
    std_ddcrm = np.append(std_ddcrm, np.std(rew_ddcrm[0:i])) 
  else:
    std_ddcrm = np.append(std_ddcrm, np.std(rew_ddcrm[i-b:i])) 

df = pd.read_csv("crm_learning_MiniGrid-Empty-6x6-v0_10000.csv")
rew_dqn = df.avg_reward.values
std_dqn = np.array([0])
for i in range(1,len(rew_dqn)):
  if i-b<0:
    std_dqn = np.append(std_dqn, np.std(rew_dqn[0:i])) 
  else:
    std_dqn = np.append(std_dqn, np.std(rew_dqn[i-b:i])) 
```

```python
steps = [i for i in range(len(rew_dcrm))]
plt.plot(steps,rew_dcrm, color="#5E35B1", label = "Q")
plt.fill_between(steps,rew_dcrm+std_dcrm, rew_dcrm-std_dcrm,
                 color="#5E35B1", edgecolor="#FFFFFF", alpha=0.2)

plt.plot(steps,rew_ddcrm, color="#40E0D0", label = "QRM")
plt.fill_between(steps,rew_ddcrm+std_dcrm, rew_ddcrm-std_ddcrm,
                 color="#40E0D0", edgecolor="#FFFFFF", alpha=0.2)

plt.plot(steps,rew_dqn, color="#FF0000", label = "CRM")
plt.fill_between(steps,rew_dqn+std_dqn, rew_dqn-std_dqn,
                 color="#FF0000", edgecolor="#FFFFFF", alpha=0.2)

plt.legend(loc=4)
plt.xlabel("Episodes")
plt.ylabel("Normalized reward")
plt.title("Empty 6x6")
```
